DAG Directed Acyclic Graph o Grafo Acíclico Dirigido

Es una estructura de datos formada por nodos conectados mediante aristas dirigidas ➡️, que representan relaciones o dependencias. 

Su característica fundamental es la ausencia total de ciclos, es decir, no existe ningún camino que empiece y termine en el mismo nodo.

Se utiliza para representar flujos, dependencias o procesos secuenciales.


Esto es un DAG en su más amplia definición:

```python
# Probar el ejemplo
from src.dag_playground import *

# Añadir relaciones (edges)
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("B", "F"),
    ("C", "F")
    
]

draw_DAG(edges)
```

En este ejemplo, podemos ver que hay nodos hoja, que serían C, D y F, además nodos intermedios como B y C y el nodo raíz de A.

Se muestra la generalización de un DAG, donde por ejemplo, el nodo hoja F, tiene dos padres, el nodo B y el C.

Pero hay diferentes tipos de DAG, uno de ellos es el árbol, y en ese caso, sigue una **estructura jerárquica** como vemos en este ejemplo:

```python
# Probar el ejemplo
from src.dag_playground import *

# Añadir relaciones (edges)
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "F")
    
]

draw_DAG(edges)
```

En un árbol ninguno nodo hoja tiene dos padres, se sigue una jerarquía

Si quieres más información detallada y ver todos los tipos de DAG, puedes ir a <https://en.wikipedia.org/wiki/Directed_acyclic_graph>.
